```python
#Importing Pandas library
import pandas as pd

#Creating a dataFrame of the text file
column_names = ['col1', 'col2', 'server_id', 'col4', 'col5', 'col6', 'col7', 'col8', 'col9', 'col10', 'col11', 'col12', 'col13', 'col14', 'col15', 'failure_time', 'col17', 'col18', 'col19', 'col20', 'col21']
error_disk_df = pd.read_csv("/content/error_disk.20160512.txt", header= None, names = column_names)
error_disk_df.head()

#Loading the size of the dataFrame
size = error_disk_df.size

#Grouping the servers which have more than 1 failures at the same time
grouped = error_disk_df.groupby("failure_time").filter(lambda x: len(x) > 1)

#Creating the dataFrame of the time values with greater than 1 server ids
failed_servers_df = grouped[["failure_time", "server_id"]]

#Showing the dataFrame
print(failed_servers_df.size)
failed_servers_df.head()

```

```python
# To group the timeValues with same serverID vallues
grouped_df = error_disk_df.groupby('failure_time')['server_id'].apply(list).reset_index(name='server_ids')

# To print the grouped values
grouped_df.head()
```
